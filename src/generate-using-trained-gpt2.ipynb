```
from os import listdir
from os.path import isfile, join
import pickle
result = {}

onlyfiles = [f for f in listdir("/content/drive/MyDrive/Yonsei/CS4-2/Sojong20Fall/newf/0.1") if isfile(join("/content/drive/MyDrive/Yonsei/CS4-2/Sojong20Fall/newf/0.1", f))]
for path in onlyfiles:
    print(path)
    pathe = "/content/drive/MyDrive/Yonsei/CS4-2/Sojong20Fall/newf/0.1/" + path
    with open(pathe, 'rb')as p:
        temp = pickle.load(p)
        result.update(temp)

with open("/content/drive/MyDrive/Yonsei/CS4-2/Sojong20Fall/newf/0.1/result.pickle", 'wb') as p:
    pickle.dump(result, p, pickle.HIGHEST_PROTOCOL)
```

```
import pickle
import pandas as pd
from itertools import islice
with open("/content/drive/MyDrive/Yonsei/CS4-2/Sojong20Fall/newf/0.1/result.pickle", 'rb') as p:
    res = pickle.load(p)


print(dict(list(res.items())[0:2]))
dataset = pd.read_csv("/content/drive/MyDrive/Yonsei/CS4-2/Sojong20Fall/newf/data_complete.csv")

for i in range(len(dataset)):
    try:
        dataset.loc[i, ['result']] = res[dataset.iloc[i]['sent_list']].replace("<|", "")
    except:
        dataset.loc[i, ['result']] = ""
        print("Error")
    print(str(i+1)+"/"+str(len(dataset)))

dataset.to_csv("/content/drive/MyDrive/Yonsei/CS4-2/Sojong20Fall/newf/0.1/result.csv", mode='w')
```
